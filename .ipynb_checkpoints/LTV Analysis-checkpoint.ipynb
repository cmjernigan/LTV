```python
import pandas as pd
from IPython.display import display

data = pd.read_csv('lss_20190531.csv')

funds = data[data['LenderCategory'] == 'FUND'].drop(['OriginationDate', 'Term', 'OriginalTerm', 'MaturityDate',
                                                      'OriginalMaturityDate', 'HasExtended', 'NumExtensions', 'Borrower',
                                                      'DaysLate', 'Address', 'City', 'State', 'Zip', 'County',
                                                      'Status', 'Substatus', 'LenderId', 'LenderCategory',
                                                      'LenderRate', 'AdvisorFee', 'ServiceFee', 'AllFees',
                                                      'EffectiveLenderRate', 'EffectiveServicingFee',
                                                      'Advisor', 'FICO', 'LTC'], axis=1).reset_index(drop=True)

funds.LoanAmount = funds.LoanAmount.str.replace('$', '').str.replace(',', '').astype(float)
funds.LoanAmountFunded = funds.LoanAmountFunded.str.replace('$', '').str.replace(',', '').astype(float)

Funds = funds.drop(['Lender', 'LenderAmount'], axis=1).drop_duplicates().reset_index(drop=True)

NCFL = funds[funds['Lender'] == 'Noble Capital Funding Line, LLC'].drop('Lender', axis=1).reset_index(drop=True)
NCIFI = funds[funds['Lender'] == 'NC Income Fund I, LLC'].drop('Lender', axis=1).reset_index(drop=True)
NCIFII = funds[funds['Lender'] == 'Noble Capital Income Fund II, LLC'].drop('Lender', axis=1).reset_index(drop=True)
NCIFIII = funds[funds['Lender'] == 'Noble Capital Income Fund III, LLC'].drop('Lender', axis=1).reset_index(drop=True)
NCPreF = funds[funds['Lender'] == 'Noble Capital Premier Fund, LLC'].drop('Lender', axis=1).reset_index(drop=True)
NCPF = funds[funds['Lender'] == 'Noble Capital Portfolio Fund, LLC'].drop('Lender', axis=1).reset_index(drop=True)
NCPFII = funds[funds['Lender'] == 'Noble Capital Portfolio Fund II, LLC'].drop('Lender', axis=1).reset_index(drop=True)
NCPFIII = funds[funds['Lender'] == 'Noble Capital Portfolio Fund III, LLC'].drop('Lender', axis=1).reset_index(drop=True)

Funds.head()
```
