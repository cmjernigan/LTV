```python
x = NCFL

import numpy as np
def geomean(a):
    b = a + 1
    return b.prod()**(1 / len(a)) - 1
def wgmean(a, w):
    b = (a + 1) ** w
    return b.prod()**(1 / w.sum()) - 1
if x.columns.isin(['LenderAmount']).max() == False:
    x['FundedWeight'] = x.LoanAmountFunded / x.LoanAmountFunded.sum()
else:
    x['FundedWeight'] = x.LenderAmount / x.LenderAmount.sum()
x['fw'] = x.LoanAmountFunded / x.LoanAmountFunded.sum()
x['lw'] = x.LoanAmount / x.LoanAmount.sum()
x['rltv'] = 1 / x.LTV
x['rltc'] = 1 / x.LTC
x['cost'] = x.LoanAmount / x.LTC
x.cost = x.cost.round(2)
medltv = x.LTV.median()
mltv = x.LTV.mean()
gmltv = geomean(x.LTV)
wgltv = wgmean(x.LTV, x.fw)

print('median:', medltv)
print('mean:', mltv)
print('geomean:', gmltv)
print('weighted geomean:', wgltv)
```
